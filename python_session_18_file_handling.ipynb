- mode

    - by defualt the mode is : **r** read mode
 
- if we want to create file and write any info then we need to change the mode
- 'r'       open for reading (default)
- 
'w'       open for writing, truncating the file firs
- 
'x'       create a new file and open it for writi
- g
'a'       open for writing, appending to the end of the file if it exi
- ts
'b'       binary 
- ode
't'       text mode (def
- ult)
'+'       open a disk file for updating (reading and writing)
